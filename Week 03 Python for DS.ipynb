```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('diabetes_model.csv')
plt.figure(figsize=(8,8))
plt.hist(data['Age'], rwidth = 0.9)
plt.title('Histogram of Age', fontsize=16)
plt.xticks()
plt.yticks()
plt.xlabel('Age') 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('diabetes_model.csv')
plt.figure(figsize=(8,8))
plt.scatter(data['BMI'], data['Glucose'], s=10, marker= '^')
plt.title('Scatter plot of BMI vs Glucose', fontsize=16)
plt.xticks()
plt.yticks()
plt.xlabel('BMI')
plt.ylabel('Glucose')

```

```python
import pandas as pd
data = pd.read_csv('diabetes_model.csv')
data.replace(to_replace=data['BMI'].min(), value=data['BMI'].mean())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('diabetes_model.csv')
sns.boxplot( x="Outcome", y="Glucose", hue='Glucose', data=data)
plt.title('Box plot of Outcome vs Glucose', fontsize=16)
plt.xticks()
plt.yticks()
plt.xlabel('Outcome')
plt.ylabel('Glucose')
```

### 5.

**-** Create a numpy array (first) with 5 x 5 size and all elements are 0 in that.

**-** Create a numpy array (second) with 5 x 5 and all elements are 1 in that.

**-** Combine the arrays first and second vertically, the final array size should be 10 x 5

```python
import numpy as np
arr1=np.zeros((5,5),dtype=int)
print(arr1)
```

```python
arr2=np.ones((5,5),dtype=int)
print(arr2)
```

```python
arr3=np.concatenate((arr1,arr2),axis=0)
print(arr3)
```
